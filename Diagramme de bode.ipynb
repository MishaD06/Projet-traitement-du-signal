```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du circuit
R = 113.08
L = 10e-3
C = 7.82e-9

# Fréquences
f = np.logspace(1, 6, 1000)
omega = 2 * np.pi * f

# Fonction de transfert
T = (1j * omega * R * C) / (1 - L * C * omega**2 + 1j * omega * R * C)

# Module et phase
gain_db = 20 * np.log10(np.abs(T))
phase = np.angle(T, deg=True)

# Tracer le diagramme de Bode
fig, ax = plt.subplots(2, 1, sharex=True)

# Gain en dB
ax[0].semilogx(f, gain_db)
ax[0].set_ylabel('Gain (dB)')
ax[0].set_title('Diagramme de Bode')
ax[0].grid(True)

# Phase
ax[1].semilogx(f, phase)
ax[1].set_ylabel('Phase (deg)')
ax[1].set_xlabel('Fréquence (Hz)')
ax[1].grid(True)

plt.show()
```
