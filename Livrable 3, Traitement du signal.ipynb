## <u>I) Contexte</u>

**Général :** L'agent K57, après une mission d'infiltration réussie dans une base secrète de cyberpirates, s'est retrouvé coincé dans une salle de conférence sans accès réseau. La composition des murs empêchait toute communication extérieure via les ondes traditionnelles. Le seul équipement disponible était un micro branché au système d'audioconférence, uniquement connecté au réseau lors des réunions. Notre mission au département R&D de l'agence AIL3C est de trouver une solution technique permettant à l'agent de communiquer vers l'extérieur dans de telles situations critiques. Nous devons proposer une solution innovante, basée sur les caractéristiques des ondes sonores, pour éviter que de telles situations ne se reproduisent.

**Dans ce Livrable 3 :** L'objectif de ce livrable est de préparer une présentation détaillée de la chaîne de transmission pour la Proof of Concept (POC) qui sera présentée au prochain comité de pilotage du département Recherche et Développement. Cette présentation doit expliquer de manière exhaustive le fonctionnement de la solution de transmission de l'information, depuis la création du son jusqu'à sa réception et son exploitation.

## <u>IV) Impact sur la chaîne de transmission</u>

**Texte court**

La chaîne de transmission reste inchangée car il s'agit d'un simple fichier de petite taille.

**Fichier texte volumineux**

L'agent active le microphone pour établir une connexion réseau. À partir de son appareil, il utilise un protocole PPP pour transmettre le fichier, qui est automatiquement segmenté en plusieurs trames, en fonction de sa taille. L'agent responsable de la surveillance du réseau reçoit ces segments, qui sont ensuite réassemblés automatiquement pour reformer le fichier original, permettant ainsi sa lecture intégrale. Cela prendra donc plus de temps.

**Fichier binaire quelconque**

La transmission d'un fichier binaire suivra la même chaîne que pour un fichier texte volumineux, avec une encapsulation dans des trames PPP.

**Son**

L'agent active le microphone et capte l'onde sonore, établissant ainsi une connexion réseau dans la salle de conférence. À l'aide de son appareil, il module le signal et le transmet dans les hautes fréquences. En utilisant la transformée de Fourier, il filtre le signal et envoie une trame PPP. L'agent extérieur reçoit cette trame, démodule le signal, applique l'inverse de la transformée de Fourier et restitue le son original.

### 3. Modulation FSK


**a) Définition**

La modulation est un processus qui consiste à faire varier une ou plusieurs caractéristiques d'un signal porteur (amplitude, fréquence ou phase) en fonction d'un autre signal appelé signal modulant. Ce processus permet de transmettre de l'information sur le signal porteur, qui peut alors être propagé sur un canal de transmission (fil, onde radio, etc.).

**b) Type de modulation (FSK)**

Nous avons choisi la modulation par déplacement de fréquence (FSK) pour les raisons suivantes :

- Robustesse au bruit : La modulation FSK est relativement robuste aux bruits et interférences sur le canal de transmission, car les variations de fréquence sont moins sensibles aux perturbations que les variations d'amplitude ou de phase.
- Faible consommation d'énergie : Les circuits FSK consomment généralement moins d'énergie que d'autres types de modulation, ce qui les rend intéressants pour les applications à faible puissance.
- Synchronisation simplifiée : La démodulation FSK ne nécessite pas de synchronisation de phase complexe entre l'émetteur et le récepteur.


La modulation FSK apparaît alors plus adaptée dans notre situation.


**c) Principe de la FSK :**


Étape 1 : Modulation de la fréquence

    Fréquences porteuses : Utilisons deux fréquences différentes pour représenter les bits 0 et 1. Par exemple, 12 kHz pour le bit 0 et 18 kHz pour le bit 1.
    Modulation : Pour chaque bit du signal Manchester, générons une onde sinusoïdale à la fréquence correspondante.
        Bit 0 : Onde sinusoïdale à 12 kHz.
        Bit 1 : Onde sinusoïdale à 18 kHz.

Étape 2 : Génération de l'onde sinusoïdale

    Onde sinusoïdale : Générons une onde sinusoïdale pour chaque bit du signal Manchester.
        Pour le bit 0 : Asin⁡(2π⋅12000⋅t)
        Pour le bit 1 : Asin⁡(2π⋅18000⋅t)
        Où A est l'amplitude et t est le temps.

Étape 3 : Concaténation des ondes

    Concaténation : Concaténons les ondes sinusoïdales pour former le signal modulé complet.

Étape 4 : Échantillonnage et Quantification

    - Fréquence d'échantillonnage : La fréquence d'échantillonnage doit être au moins deux fois la fréquence maximale du signal pour éviter l'aliasing. Si la fréquence maximale est 18 kHz, nous devons choisir une fréquence d'échantillonnage d'au moins 36 kHz. Pour une meilleure qualité, nous pouvons choisir une fréquence d'échantillonnage plus élevée, par exemple 44.1 kHz.
    
    - Échantillons : Prenons des échantillons du signal modulé à intervalles réguliers déterminés par la fréquence d'échantillonnage.
    
    - Niveaux de quantification : Divisons l'amplitude du signal en plusieurs niveaux discrets. Par exemple, pour une quantification sur 8 bits, il y aura 256 niveaux.
    
    - Quantification : Assignons chaque échantillon à l'un des niveaux de quantification les plus proches.

### 5. Trame PPP

PPP (Point-to-Point Protocol) est un protocole de communication utilisé pour établir une connexion directe entre deux nœuds. Il est couramment utilisé pour les connexions Internet par modem, les liaisons série, et d'autres types de connexions point à point. PPP permet de transporter plusieurs types de trafic réseau sur une liaison série, y compris IP, IPX, et AppleTalk.

Encapsulation PPP

    Trame PPP : Encapsulons le signal numérique enregistré dans une trame PPP. Une trame PPP typique commence par un drapeau (0x7E), suivi de l'adresse, du contrôle, des données, et se termine par un autre drapeau (0x7E).

Exemple de trame PPP :

    Drapeau : 0x7E
    Adresse : 0xFF (adresse de diffusion)
    Contrôle : 0x03 (contrôle non numéroté)
    Données : Signal numérique enregistré
    Drapeau : 0x7E
